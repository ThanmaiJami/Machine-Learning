```python
#matplotlib
#25/01/2022
import matplotlib.pyplot as plt
import numpy as np

#day one, the numbers and marks of B9 students:
x = np.array([10,12,14,15,16,17,21,22,23,24,26,29,25])
y = np.array([99,86,87,88,99,86,93,87,94,78,77,85,86])

plt.scatter(x, y,color='blue')
#day two, the numbers and marks of B14 students:
x = np.array([13,25,32,36,45])
y = np.array([85,96,77,82,96])
plt.scatter(x, y,color='yellow')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'solid')   #dashed dotted dashdot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#day one, the numbers and marks of B16 students:
x = np.array([10,12,14,15,16,17,21,22,23,24,26,29,25])
y = np.array([99,86,87,88,99,86,93,87,94,78,77,85,86])

plt.plot(x,linestyle = 'dashed')
plt.plot(y,linestyle = 'dashed')
#day two, the numbers and marks of B17 students:
x = np.array([13,25,32,36,45])
y = np.array([85,96,77,82,96])
plt.plot(x,linestyle = 'dashed')
plt.plot(y,linestyle = 'dashed')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "yellow", "red"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```
